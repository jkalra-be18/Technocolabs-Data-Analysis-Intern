```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
covid_data = pd.read_csv(r'C:\Users\Administrator\Desktop\covid.csv', encoding = 'latin', low_memory= False)
covid_data.head()
```

```python
covid_data.shape
```

```python
covid_final = pd.DataFrame()
covid_final = covid_data.dropna()
```

```python
covid_final = covid_final.set_index('Country')
covid_final.tail()
```

```python
eei = pd.DataFrame()
eei = covid_final.sort_values(['Covid 19 Economic exposure index'], ascending = False)
eei.head()
```

```python
x = pd.DataFrame()
x = eei.head()
sns.barplot(x.index, x['Covid 19 Economic exposure index'])
plt.xlabel('Countries', size = 10)
plt.ylabel('Economic Exposure Index')
plt.title('Top 5 Countries with highest EEI')
plt.show()
```

```python
td = covid_final.sort_values(['tourism as percentage of GDP'], ascending = False)
td.head()
```

```python
y = pd.DataFrame()
y = td.head()
sns.barplot(y.index, y['tourism as percentage of GDP'])
plt.xlabel('Countries', size = 10)
plt.ylabel('Tourism % in GDP')
plt.title('Top 5 Countries with highest percentage of gdp from Tourism')
plt.xticks(rotation = 45)
plt.show()
```

```python
fid = covid_final.sort_values(['food import dependence '], ascending = True)
fid.head()
```

```python
z = pd.DataFrame()
z = fid.head()
sns.barplot(z.index, z['food import dependence '])
plt.xlabel('Countries', size = 10)
plt.ylabel('Food import Dependence')
plt.title('Top 5 Countries with lowest Food Import Dependence')
plt.xticks(rotation = 45)
plt.show()
```

```python
ggd = covid_final.sort_values(['Tourism dependence'], ascending = True)
ggd.head()
```

```python
a = pd.DataFrame()
a = ggd.head()
sns.barplot(x = a.index, y = a['Tourism dependence'])
plt.xlabel('Countries', size = 10)
plt.ylabel('Tourism Dependence')
plt.title('Top 5 Countries with lowest Tourism Dependence')
plt.xticks(rotation = 45)
plt.show()
```
