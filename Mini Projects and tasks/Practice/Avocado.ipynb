```python
df = pd.read_csv(r'C:\\Users\\Administrator\\Desktop\\avocado.csv')
df = df.copy()[df['type']=='organic']

df["Date"] = pd.to_datetime(df["Date"])

df.sort_values(by="Date", ascending=True, inplace=True)
df.head()


```

```python
graph_df = pd.DataFrame()
for region in df['region'].unique()[:16]:
    region_df = df.copy()[df['region']==region]
    region_df.set_index('Date',inplace = True)
    region_df.sort_index(inplace = True)
    region_df[f"{region}_New"]=region_df["AveragePrice"].rolling(25).mean()
    if graph_df.empty:
        graph_df = region_df[[f"{region}_New"]]  # note the double square brackets!
    else:
        graph_df = graph_df.join(region_df[f"{region}_New"])
graph_df.tail()
```

```python
graph_df.plot(figsize=(8,5), legend=False)
```
